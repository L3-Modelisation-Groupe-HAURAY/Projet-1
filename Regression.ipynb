```python
import numpy as np; import matplotlib.pyplot as plt
```

```python
t=np.array([0, 0.5,1,1.5,2,3,5,7.5,10,15,20,24],dtype='float')
```

```python
y=np.array([23.8,22.1,20.5,19,17.6,15.2,11.3,7.82,5.39,2.57,1.22,0.68],dtype='float')
```

```python
lny=np.log(y)
```

```python
from scipy.stats import linregress
```

```python
(a,b,rho,p,stderr)=linregress(t,lny)
```

```python
y_th=np.exp(b+a*t)
```

```python
plt.plot(t,y,'Dr',label='Données expérimentales') ; 
plt.plot(t,y_th,label='régression "exponentielle"') ; 
plt.legend(); 
plt.title('Concentration en lipoamide au cours du temps'); 
plt.show()
```

```python
t2=np.arange(0,44); print(t2)
```

```python
y2 =np.array([2313,2653,2548,3499,3593,3235,3527,4208,5324,5988,6554,5560,4783,5240,5198,6202,5907,5973,5215,
              4047,4053,4783,4669,4585,4807,4318,3599,3037,3834,4204,3950,4697,4094,3153,2973,2666,3787,3494,
              3492,3047,2251,2727,3370,2644],dtype='float')
```

```python
A=np.ones([44,3]); 
for i in t2 :
    A[i] = [1,i,i*i]; 
B = A.T ; C= B.dot(A); print(C);
z2=B.dot(y2);
(a2,b2,c2)=np.linalg.solve(C,z2); print(a2,b2,c2);
```

```python
A3=np.ones([44,4]); 
for i in t2 :
    A3[i] = [1,i,i*i,i*i*i]; 
B3 = A3.T ; C3= B3.dot(A3); print(C3);
z3=B3.dot(y2);
(a3,b3,c3,d3)=np.linalg.solve(C3,z3); print(a3,b3,c3,d3);
```

```python
A4=np.ones([44,5]); 
for i in t2 :
    A4[i] = [1,i,i*i,i*i*i,i*i*i*i]; 
B4 = A4.T ; C4= B4.dot(A4); print(C4);
z4=B4.dot(y2);
(a4,b4,c4,d4,e4)=np.linalg.solve(C4,z4); print(a4,b4,c4,d4,e4);
```

```python
t3=np.arange(0,55); y3=np.zeros(55);
y2_th= a2+ b2*t3+c2*t3*t3 ;
y3_th= a3+ b3*t3+c3*t3*t3 + d3*t3*t3*t3;
y4_th= a4+ b4*t3+c4*t3*t3 + d4*t3*t3*t3 + e4*t3*t3*t3*t3;
```

```python
plt.bar(t2,y2,label='Cas déclarés'); 
plt.plot(t3,y2_th,'r',label='Interpolation de degré 2') ; 
plt.plot(t3,y3_th,'g',label='Interpolation de degré 3') ;
plt.plot(t3,y4_th,'y',label='Interpolation de degré 4') ;
plt.legend(); 
plt.title("Cas journaliers de COVID-19 en Italie et interpolation d'ordre 2/3/4" );
fig = plt.gcf(); fig.set_size_inches(9,6);
plt.show()
```
